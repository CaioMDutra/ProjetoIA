```python
# Passo 1 - Importar base de dados
import pandas as pd

tabela = pd.read_excel("base_churn.xlsx")

#verificar se tem algum valor vazio
display(tabela.info())
```

```python
# Passo 2 - Preparar a base de dados para a inteligência artifical
# Label Encoder
from sklearn.preprocessing import LabelEncoder

codificador = LabelEncoder()
tabela["FCR"] = codificador.fit_transform(tabela["FCR"])
tabela["SLA"] = codificador.fit_transform(tabela["SLA"])
tabela["CSAT"] = codificador.fit_transform(tabela["CSAT"])
display(tabela.info())
  # profissao
    # cientista -> 0
    # professor -> 1
    # pintor -> 2
    # médico -> 3


# mix_credito
# comportamento_pagamento
```

```python
#dados de x (dados para previsão) e de y (a previsão)
y = tabela["Situação"]
colunas = ["Situação"]
x = tabela.drop(columns=colunas)

# treino e teste
from sklearn.model_selection import train_test_split

x_treino, x_teste, y_treino, y_teste = train_test_split (x, y, test_size=0.3)
```

```python
# Passo 3 - Criar o modelo de IA -> Score de crédito: Good (Bom), Standard (Médio), Poor (Ruim)
# arvore de decisão
from sklearn.ensemble import RandomForestClassifier
# knn (vizinhos mais próximos)
from sklearn.neighbors import KNeighborsClassifier

modelo_arvoredecisao = RandomForestClassifier()
modelo_knn = KNeighborsClassifier()

modelo_arvoredecisao.fit(x_treino, y_treino)
modelo_knn.fit(x_treino, y_treino)
```

```python
# Passo 4 - Escolher o modelo
from sklearn.metrics import accuracy_score

previsao_arvoredecisao = modelo_arvoredecisao.predict(x_teste)
previsao_knn = modelo_knn.predict(x_teste.to_numpy())

print(accuracy_score(y_teste, previsao_arvoredecisao))
print(accuracy_score(y_teste, previsao_knn))
```
